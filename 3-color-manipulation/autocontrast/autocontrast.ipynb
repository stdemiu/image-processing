```python
from skimage.io import imread, imshow, imsave

img = imread('image-low-contrast.png')

img.size
img.min()

imshow(img);

img_f = img.astype('float')
img_f.dtype

k = round(img_f.size * 0.05)
k

pixels = img_f.ravel().copy()
pixels.sort()

pixels

min_p, max_p = pixels[k + 1], pixels[-(k + 1)]
min_p, max_p

img_contrast = (img_f - min_p) * (255 / (max_p - min_p))
img_contrast

import numpy as np

img_contrast = np.clip(img_contrast, 0, 255)
img_final = img_contrast.astype('uint8')

imshow(img_final)
img_stable = imread('image-gray.png')
np.array_equal(img_stable, img_final)
```

```python
from skimage.io import imread, imsave
from skimage.color import rgb2gray
import numpy as np

img = imread('image.jpg')   

if img.ndim == 3:
    img_gray = rgb2gray(img)
else:
    img_gray = img / 255.0

img_gray = (img_gray * 255).astype(np.uint8)
img_low = np.clip((img_gray / 2) + 60, 0, 255).astype(np.uint8)
img_mid = img_gray
img_high = np.clip((img_gray - img_gray.min()) * (255 / (img_gray.max() - img_gray.min())), 0, 255).astype(np.uint8)


imsave('image-low-contrast.png', img_low)
imsave('image-gray.png', img_mid)
imsave('image-high-contrast.png', img_high)

```
