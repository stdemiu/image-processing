```python
from skimage.io import imread, imsave, imshow

img = imread('image.jpg')
img.dtype  

from skimage import img_as_float

img_f = img_as_float(img)  

img_f.dtype
img.min(), img.max()
img_f.min(), img_f.max()

imshow(img_f / 1.5); 
imshow(img_f / 2);
imshow(img_f / 4);
imshow(img_f / 4 + 0.25);

from numpy import clip  

imshow(clip(img_f * 1.5, 0 , 1));
imshow(clip(img_f * 2, 0 , 1));
imshow(clip(img_f * 2 - 0.5, 0, 1));
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread("image.jpg")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
brightness = np.mean(gray)
print(brightness)

plt.imshow(gray, cmap="gray")
plt.axis("off")
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread("image.jpg")
negative = 255 - img

plt.imshow(cv2.cvtColor(negative, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread("image.jpg")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
avg_brightness = np.mean(gray)
print(avg_brightness)

plt.imshow(gray, cmap="gray")
plt.axis("off")
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread("image.jpg")
b, g, r = cv2.split(img)

plt.figure(figsize=(10,4))
plt.subplot(131); plt.imshow(b, cmap="gray"); plt.axis("off")
plt.subplot(132); plt.imshow(g, cmap="gray"); plt.axis("off")
plt.subplot(133); plt.imshow(r, cmap="gray"); plt.axis("off")
plt.show()
```
